```python
# Importing Image class from PIL module
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
 
prefix = '/content/drive/MyDrive/CS236/CS236 Final Project/'
# Opens a image in RGB mode
im = Image.open(r"/content/drive/MyDrive/CS236/CS236 Final Project/WassersteinGAN/WGAN_samples_4/fake_samples_28500.png")
 
# Size of the image in pixels (size of original image)
# (This is not mandatory)
width, height = im.size
 
# Setting the points for cropped image

epoch = '28500'

for i in range(8):
  for j in range(8):
      left = 64 * j
      top = 64 * i
      right = 64 * (j + 1)
      bottom = 64 * (i + 1)
      # Cropped image of above dimension
      # (It will not change original image)
      im1 = im.crop((left, top, right, bottom))
      imgplot = plt.imshow(im1)
      plt.axis('off')
      print(im1.size)
      print(os.path.join(prefix, f'wgan_out/{epoch}_{i*8 + j}.png'))
      im1.save(os.path.join(prefix, f'wgan_out/{epoch}_{i*8 + j}.png'))
      plt.show()
```
